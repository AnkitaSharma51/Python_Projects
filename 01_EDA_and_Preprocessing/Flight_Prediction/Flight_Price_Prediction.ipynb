## Feature Engineering Process

```python
# Splitting of Date_of_Journey
df['Day']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df['Day']=df['Day'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
df.info()
```

```python
df.head()
```

```python
df['Arrival_Time']
```

```python
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
df['Arrival_Time']
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Mins']=df['Arrival_Time'].str.split(':').str[1]
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_Mins']=df['Arrival_Mins'].astype(int)
```

```python
df.info()
```

```python
df.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Mins']=df['Dep_Time'].str.split(':').str[1]
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_Mins']=df['Dep_Mins'].astype(int)
df.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
# Using Total_Stops
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
df.head()
```

```python
df.info()
```

```python
# Converting Duration into Minutes
df['Duration'].head()
```

```python
df['Duration_Hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df[df['Duration_Hour']=='5m']
```

```python
df.drop(6474,axis=0,inplace=True)
df.drop(13343,axis=0,inplace=True)
```

```python
df['Duration_Hour']=df['Duration_Hour'].astype(int)
```

```python
df['Duration_minutes']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Duration_minutes'].replace(np.nan,0,inplace=True)
```

```python
df['Duration_minutes']=df['Duration_minutes'].astype(int)
```

```python
df['Duration_minutes'].isnull().sum()
```

```python
df["Duration_in_minutes"]=df['Duration_Hour'].astype(int)*60+df['Duration_minutes'].astype(int)
```

```python
df.drop(['Duration'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Airline']=le.fit_transform(df['Airline'])
df['Source']=le.fit_transform(df['Source'])
df['Destination']=le.fit_transform(df['Destination'])
df['Additional_Info']=le.fit_transform(df['Additional_Info'])
```
